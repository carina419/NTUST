```python
import numpy as np
x_data = [1.0, 2.0, 3.0, 4.0]
y_data = [3.0, 10.0, 21.0, 36.0]

w1 = 2.0;  w2= 1.0  # a random guess: random value
w=np.array([w1, w2])
# our model forward pass
def forward(x, w1, w2):
   return x * w1+(x**2)*w2


# Loss function
def loss(x, y,w1, w2):
   y_pred = forward(x, w1, w2)
   return (y_pred - y) * (y_pred - y)


# compute gradient
def gradient(x, y, w1, w2):  # d_loss/d_w
   return np.array([2 * x * ((x**2 )* w2 +x*w1- y), 2*(x**2)*((x**2 )* w2 +x*w1- y)])
```

```python
# Before training
print("predict (before training)",  35, forward(5,2,1))
w1=2; w2=1
w=np.array([w1, w2])
# Training loop
for epoch in range(1000):
   for x_val, y_val in zip(x_data, y_data):
       grad = gradient(x_val, y_val, w1, w2)
       w = w - 0.01 * grad
       print("\tgrad: ", x_val, y_val, grad, w)
       w1=w[0]; w2=w[1]
       l = loss(x_val, y_val, w1, w2)

   print("progress:", epoch, "w=", w[0],w[1], "loss=", l)

# After training
print("predict (after training)",  "4 hours", forward(5,w[0], w[1]))
```
