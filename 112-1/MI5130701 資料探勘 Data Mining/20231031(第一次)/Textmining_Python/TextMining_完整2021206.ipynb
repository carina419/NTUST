# Text Preprocessing  文本預處理 

由於取得的文本資料不見得是很乾淨，像是利用爬蟲(Web Crawler)擷取下來的內容中有很多是沒必要的標籤，  
或者是文本內有一些特殊的非英文字和特殊符號等，這些都是對文本分析沒有太大幫助的，可以事先去除掉。

這樣的程序可以直接用 Python 的正則表達式套件 re 來處理，re 是 regular expression 的縮寫

## 找出符合的關鍵字的句子

re.**search**(pattern, string, flags=0)  
從 string 中找尋第一個配對形式字串 pattern ，找到回傳配對物件，沒有找到回傳 None 

re.**match**(pattern, string, flags = 0)  
判斷配對形式字串 pattern 是否與 string 的開頭相符，如果相符就回傳配對物件，不相符就回傳 None 

- pattern : 正則中的模式字串
- string : 要被查詢替換的原始字串
- flags : 用於控制正則表示式的匹配方式，如：是否區分大小寫，多行匹配等等

## 去除文本中的非文字部份

re.sub(pattern, repl, string, count = 0, flags = 0)  
依據 pattern 及 repl 對 string 進行處理，結果回傳處理過的新字串

共有五個參數

- 其中三個必填參數：pattern, repl, string

- 兩個可選參數：count, flags

```python
# 去除括號和括號内的所有內容
rt =  "\\【.*?】+|\\《.*?》+|\\#.*?#+|[.!/_,$&%^*()<>+""'?@|:~{}#]+|[——！\\\，。=？、：“”‘’￥……（）《》【】]"
```

## NLTK 套件也有斷詞功能喔! 請使用tokenize()函數

一般我們可以透過 split() ，先做到簡單的斷詞，看一個簡單的英文範例：

```python
# 英文範例
sentence = """At eight o'clock on Thursday morning
... Arthur didn't feel very good."""
# 使用 split
sentence.split()
```

各位有發現不完美的地方嗎?有發現到嗎？  

上述斷詞後的結果出現 didn't、 good. 的字詞，分別在英文縮寫以及帶有句點的地方無法斷開。  

原因為 split 僅能利用空格做為分隔符號進行斷詞。而 NLTK 提供了 word_tokenize() 的功能可以解決此問題。

在使用NLTK執行分詞之前，我們需要先安裝「punkt」套件。「punkt」包含了許多預訓練好的分詞模型。  
如果沒有安裝「punkt」，我們在使用時系統將會報錯，提示我們進行安裝。參考網站  
https://kknews.cc/zh-tw/tech/eye8qrq.html

```python
# 使用 tokenize 進行斷詞
# nltk.download('punkt')
nltk.word_tokenize(sentence)
```

## 英文的特殊處理-1：stemming 和 lemmatization

處理英文文本資料還有個地方跟中文不同，就是要進行**詞幹提取（stemming）**和 **詞形還原（lemmatization）**

兩者的主要目的是要找到單詞的原始形式


- 詞幹提取是一個將詞語簡化為詞幹、詞根或詞形的過程（如 books-book，looked-look）

- 詞形還原的目的，詞形還原就是去掉單詞的詞綴，提取單詞的主幹部分，通常提取後的單詞會是字典中的單詞。方法多半是使用詞彙知識庫來取得正確的單詞形式。不同於詞幹提取（stemming），提取後的單詞不一定會出現在單詞中。比如，單詞“cars”詞形還原後的單詞為“car”，單詞“ate”詞形還原後的單詞為“eat”

## TF-IDF（Term Frequency - Inverse Document Frequency）

該方法是利用詞頻的概念，將文本轉換成向量，而不考慮詞的順序

假設有 N 篇文檔，在其中一篇文檔 D 中，單詞 x 的 TF、IDF、TF-IDF 分別定義如下：

- TF : Term Frequency
   指單詞 x 在目前的文檔 D 中的詞頻
   
- IDF : Inverse Document Frequency  
   N 表是在語料庫中所有文檔的總數，而 N(x) 表示在語料庫中包含單詞 x 的文檔總數

   $ IDF(x) = log\frac{N}{N(x) } $
   
   
- TF-IDF  
   $TF-DIF(x) = TF(x) * IDF(x)$

同時進行詞頻統計並計算 TF-IDF

```python
docs = np.array([
        'The sun is shining',
        'The weather is sweet',
        'The sun is shining, the weather is sweet, and one and one is two'])
docs
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(docs))
```

```python
word = vectorizer.get_feature_names()
weight = tfidf.toarray()
```

如果自己依照上面公式計算，會發現與 sklearn 的結果不同，這是因為 sklearn 的 TfidfsVectorizer 同時會進行向量化、TF-IDF 和標準化三步驟

另外在 IDF 的計算也有些不同

   $ IDF(x) = log\frac{N + 1}{1 + N(x)} $
   
   
TF-IDF的計算也稍微有些不同  

   $TF-DIF(x) = TF(x) * (IDF(x) + 1)$
   
最後結果會再做 L2 正規化

例如：第三個文件中的 is 計算內容如下

$ TF = 3 $

$ IDF = log\frac{1 + 3}{1 + 3} = 0$

TF-IDF = 3 * (0 + 1) = 3

第三份文件中每個單詞的 TF-IDF  
[3.39, 3.0, 3.39, 1.29, 1.29, 1.29, 2.0, 1.69, 1.29]


進行 L2 正規化

TF-IDF(D3, is) = 3 / (3.39^2 + 3.0^2 + 3.39^2 + 1.29^2 + 1.29^2 + 1.29^2 + 2.0^2 + 1.69^2 + 1.29^2)^(1/2) = 0.45

## N-gram 語言模型

由於詞袋模型是不考慮每個單詞在句子中的順序，所以在某些場景下，把一句話的順序搞亂，該句話會變得不知道是什麼意義

狗咬我 = 我咬狗 ??

N-gram 本身指一個由 N 個單詞組成的集合，每個單詞具有先後順序，且不要求單詞之間互不相同。常用的有 Bi-gram (N=2, N=2) 和 Tri-gram (N=3, N=3)

例如：I love deep learning，Bi-gram 和 Tri-gram 結果分別如下：

- Bi-gram : {I, love}, {love, deep}, {deep, learning}
- Tri-gram : {I, love, deep}, {love, deep, learning}

sklearn 中的 CountVectorizer 有一個參數 ngram_range 可以給予設定值，如果給值為 (2, 2) 則為 Bigram

要注意的是，使用這樣的語言模型，會增加處理的字典大小

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import jieba
```

```python
content = ['為了聯盟，為了勝利，打倒外國人！打倒外國人！',
               '朋友：為了聯盟？',
               '我是玩遊戲，玩得很投入']
```

```python
# 利用 jieba 分詞，並用 ' ' 連接
data = [' '.join(jieba.cut(e)) for e in content] 
data
```

```python
# ngram_range = (2,2) 表示 bigram，其他 ram_range = (1, 1) 表示 unigram, ngram_range = (3, 3) 表示 thirgram
vector = CountVectorizer(min_df = 1, ngram_range = (2, 2)) 
```

```python
# 將分詞好的文本轉換為矩陣
X = vector.fit_transform(data) 
```

```python
# 取得特徵組合單詞
vector.vocabulary_ 
```

## Word2vec 詞向量

Ｗord2Vec 是使用文件集的詞語去訓練模型，該模型是一個淺層的網路模型，將文章中的詞映射到一個固定長度（可自行設定該長度）的連續向量。一般使用的維度，通常為 100 ~ 500。利用單詞的向量，就可以用來判斷兩詞之間是否具有意義，而意義相近的詞之間的向量距離較小

Word2Vec 有兩種模式：

- CBOW（Continuous Bag-Of-Words）：利用單詞的上下文預測當前的詞
- Skip-Gram：利用當前的單詞來預測上下文

https://code.google.com/archive/p/word2vec/

因為 Word2Vec 模型最後得到的是詞向量，如果要表示成句子的話，最簡單的方法就是，將每個句子中的詞向量相加取平均值，即每個句子的平均詞向量來表示成句子的向量

```python
from gensim.models import Word2Vec                  
import numpy as np
```

```python
data = [["I", "love", "deep", "learning"], ["I", "love", "studying"], ["I", "want", "to", "travel"]]
```

```python
type(data)
```

```python
train_w2v = Word2Vec(data, min_count = 1, size = 50, workers = 4)
# 參數說明：
# min_count 表示單詞的詞頻小於該值會被捨棄
# size 表示特徵向量的維度
# workers 控制訓練時 CPU 並行數
```

計算平均詞向量來表示成句子的向量

```python
avg_data = []

for row in data:  
    vec = np.zeros(50)
    count = 0
    
    for word in row:
        try:
            vec += train_w2v[word]
            count += 1
        except:
            pass
    
    avg_data.append(vec / count) 
```

# 主題模型 Topic Model

主題模型主要是將未有標記標籤的文本資料，給予主題，以便可以進行後續使用。

可以將主題模型視為一個 Clustering（非監督式學習）的一種

此處將介紹一種方法，稱為 **LDA（Latent Dirichlet allocation）**

LDA 該模型認為每篇文件都是由少數幾個「主題 (Topic)」所組成，而且每個主題都可以由少數幾個重要的「單詞 (Word)」來描述
