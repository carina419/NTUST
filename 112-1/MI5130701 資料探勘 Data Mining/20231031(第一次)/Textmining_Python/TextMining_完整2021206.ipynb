å¦å¤–åœ¨ IDF çš„è¨ˆç®—ä¹Ÿæœ‰äº›ä¸åŒ

   $ IDF(x) = log\frac{N + 1}{1 + N(x)} $
   
   
TF-IDFçš„è¨ˆç®—ä¹Ÿç¨å¾®æœ‰äº›ä¸åŒ  

   $TF-DIF(x) = TF(x) * (IDF(x) + 1)$
   
æœ€å¾Œçµæœæœƒå†åš L2 æ­£è¦åŒ–

ä¾‹å¦‚ï¼šç¬¬ä¸‰å€‹æ–‡ä»¶ä¸­çš„ is è¨ˆç®—å…§å®¹å¦‚ä¸‹

$ TF = 3 $

$ IDF = log\frac{1 + 3}{1 + 3} = 0$

TF-IDF = 3 * (0 + 1) = 3

ç¬¬ä¸‰ä»½æ–‡ä»¶ä¸­æ¯å€‹å–®è©çš„ TF-IDF  
[3.39, 3.0, 3.39, 1.29, 1.29, 1.29, 2.0, 1.69, 1.29]

TF-IDF(D3, is) = 3 / (3.39^2 + 3.0^2 + 3.39^2 + 1.29^2 + 1.29^2 + 1.29^2 + 2.0^2 + 1.69^2 + 1.29^2)^(1/2) = 0.45

## N-gram èªè¨€æ¨¡å‹

ç”±æ–¼è©è¢‹æ¨¡å‹æ˜¯ä¸è€ƒæ…®æ¯å€‹å–®è©åœ¨å¥å­ä¸­çš„é †åºï¼Œæ‰€ä»¥åœ¨æŸäº›å ´æ™¯ä¸‹ï¼ŒæŠŠä¸€å¥è©±çš„é †åºæäº‚ï¼Œè©²å¥è©±æœƒè®Šå¾—ä¸çŸ¥é“æ˜¯ä»€éº¼æ„ç¾©

N-gram æœ¬èº«æŒ‡ä¸€å€‹ç”± N å€‹å–®è©çµ„æˆçš„é›†åˆï¼Œæ¯å€‹å–®è©å…·æœ‰å…ˆå¾Œé †åºï¼Œä¸”ä¸è¦æ±‚å–®è©ä¹‹é–“äº’ä¸ç›¸åŒã€‚å¸¸ç”¨çš„æœ‰ Bi-gram (N=2, N=2) å’Œ Tri-gram (N=3, N=3)

ä¾‹å¦‚ï¼šI love deep learningï¼ŒBi-gram å’Œ Tri-gram çµæœåˆ†åˆ¥å¦‚ä¸‹ï¼š

- Bi-gram : {I, love}, {love, deep}, {deep, learning}
- Tri-gram : {I, love, deep}, {love, deep, learning}

sklearn ä¸­çš„ CountVectorizer æœ‰ä¸€å€‹åƒæ•¸ ngram_range å¯ä»¥çµ¦äºˆè¨­å®šå€¼ï¼Œå¦‚æœçµ¦å€¼ç‚º (2, 2) å‰‡ç‚º Bigram

è¦æ³¨æ„çš„æ˜¯ï¼Œä½¿ç”¨é€™æ¨£çš„èªè¨€æ¨¡å‹ï¼Œæœƒå¢åŠ è™•ç†çš„å­—å…¸å¤§å°

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import jieba
```

```python
content = ['ç‚ºäº†è¯ç›Ÿï¼Œç‚ºäº†å‹åˆ©ï¼Œæ‰“å€’å¤–åœ‹äººï¼æ‰“å€’å¤–åœ‹äººï¼',
               'æœ‹å‹ï¼šç‚ºäº†è¯ç›Ÿï¼Ÿ',
               'æˆ‘æ˜¯ç©éŠæˆ²ï¼Œç©å¾—å¾ˆæŠ•å…¥']
```

```python
# åˆ©ç”¨ jieba åˆ†è©ï¼Œä¸¦ç”¨ ' ' é€£æ¥
data = [' '.join(jieba.cut(e)) for e in content] 
data
```

```python
# ngram_range = (2,2) è¡¨ç¤º bigramï¼Œå…¶ä»– ram_range = (1, 1) è¡¨ç¤º unigram, ngram_range = (3, 3) è¡¨ç¤º thirgram
vector = CountVectorizer(min_df = 1, ngram_range = (2, 2)) 
```

```python
# å°‡åˆ†è©å¥½çš„æ–‡æœ¬è½‰æ›ç‚ºçŸ©é™£
X = vector.fit_transform(data) 
```

```python
# å–å¾—ç‰¹å¾µçµ„åˆå–®è©
vector.vocabulary_ 
```

```python
# (æ–‡æª”åºåˆ—, ç‰¹å¾µçµ„åˆå–®è©)   é »ç‡
print(X)    
```

```python
from gensim.models import Word2Vec                  
import numpy as np
```

```python
data = [["I", "love", "deep", "learning"], ["I", "love", "studying"], ["I", "want", "to", "travel"]]
```

```python
type(data)
```

```python
train_w2v = Word2Vec(data, min_count = 1, size = 50, workers = 4)
# åƒæ•¸èªªæ˜ï¼š
# min_count è¡¨ç¤ºå–®è©çš„è©é »å°æ–¼è©²å€¼æœƒè¢«æ¨æ£„
# size è¡¨ç¤ºç‰¹å¾µå‘é‡çš„ç¶­åº¦
# workers æ§åˆ¶è¨“ç·´æ™‚ CPU ä¸¦è¡Œæ•¸
```

```python
print(train_w2v)
```

```python
train_w2v.wv.vocab
```

```python
train_w2v.wv.get_vector('love')
```

```python
train_w2v.wv.similar_by_word('love')
```

è¨ˆç®—å¹³å‡è©å‘é‡ä¾†è¡¨ç¤ºæˆå¥å­çš„å‘é‡

```python
avg_data = []

for row in data:  
    vec = np.zeros(50)
    count = 0
    
    for word in row:
        try:
            vec += train_w2v[word]
            count += 1
        except:
            pass
    
    avg_data.append(vec / count) 
```
