## TF-IDF（Term Frequency - Inverse Document Frequency）

like DTM, TF-IDF transforms a document into a vector of weights. The order of terms in a document is not considered.

assume that there are N documents, for a term x in document D, the TF-IDF score is calculated as follows:

- TF : Term Frequency
   number of occurrance of term x in document D.
   
- IDF : Inverse Document Frequency  
   N is the number of documents in corpus, and N(x) is the nyumber of document in corpus that contain term x.

   $ IDF(x) = log\frac{N}{N(x) } $
   
   
- TF-IDF  
   $TF-DIF(x) = TF(x) * IDF(x)$

同時進行詞頻統計並計算 TF-IDF

```python
docs = np.array([
        'The sun is shining',
        'The weather is sweet',
        'The sun is shining, the weather is sweet, and one and one is two'])
docs
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(docs))
```

TfidfsVectorizer of sklearn uses different equations and performs L2 normalization.

 IDF

   $ IDF(x) = log\frac{N + 1}{1 + N(x)} $
   
   
TF-IDF  

   $TF-DIF(x) = TF(x) * (IDF(x) + 1)$
   
 L2 normalization:

For example, the term 'is' in the third document.

$ TF = 3 $

$ IDF = log\frac{1 + 3}{1 + 3} = 0$

TF-IDF = 3 * (0 + 1) = 3

The TF-IDF of the terms in document 3 before L2 normalization:  
[3.39, 3.0, 3.39, 1.29, 1.29, 1.29, 2.0, 1.69, 1.29]


 Perform L2 normalization:

TF-IDF(D3, is) = 3 / (3.39^2 + 3.0^2 + 3.39^2 + 1.29^2 + 1.29^2 + 1.29^2 + 2.0^2 + 1.69^2 + 1.29^2)^(1/2) = 0.45

sklearn.decomposition.LatentDirichletAllocation(n_components=10, *, doc_topic_prior=None, topic_word_prior=None, learning_method='batch', learning_decay=0.7, learning_offset=10.0, max_iter=10, batch_size=128, evaluate_every=-1, total_samples=1000000.0, perp_tol=0.1, mean_change_tol=0.001, max_doc_update_iter=100, n_jobs=None, verbose=0, random_state=None)

```python
print(df.shape)
df['spam'] = df['spam'].astype(int)
df = df.drop_duplicates()   # 刪除重複的
df = df.reset_index(inplace = False)[['text','spam']]
print(df.shape)
```

```python
clean_desc = []

for w in range(len(df.text)):
    desc = df['text'][w].lower()
    #remove punctuation
    desc = re.sub('[^a-zA-Z]', ' ', desc)
    #remove tags
    desc = re.sub("</?.*?>"," <> ",desc)
    #remove digits and special chars
    desc = re.sub("(\d|\W)+"," ",desc)
    clean_desc.append(desc)

#assign the cleaned descriptions to the data frame
df['text'] = clean_desc
df.head(3)
```
