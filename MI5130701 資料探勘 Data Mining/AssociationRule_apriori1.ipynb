```python
import mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(mylist).transform(mylist)
te_ary
```

```python
te_ary.astype('int')
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support = 0.3)
```

```python
frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['length'] >= 2) & (frequent_itemsets['support'] >= 0.3)]
```

```python
frequent_itemsets[frequent_itemsets['itemsets'] == {'Nike', 'Lacoste'}]
```

```python
fname = 'Nike'
frequent_itemsets[frequent_itemsets['itemsets'].astype(str).str.title().str.contains(fname)]
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.7)
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.2)
rules
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.3)
rules
```

```python
rules['antecedent_len'] = rules['antecedents'].apply(lambda x:len(x))
rules
```

```python
rules[(rules['antecedent_len'] >= 1) & (rules['confidence'] >= 0.7) & (rules['lift'] >= 1.2)]
```

```python
sc = []
data = []

for i, j in zip(original_data["shopping_cart"], original_data["goods"]):
    if i not in sc:
        sc.append(i)
        data.append([])
    data[sc.index(i)].append(j)

data
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
te_ary
```

```python
te.columns_
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support = 0.01)
```

```python
frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.1)
```
