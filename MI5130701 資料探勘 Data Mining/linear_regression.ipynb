```python
#%% read from csv file
df = pd.read_csv('cereals.csv')  
print(df) 
#%%
####################################################
## A scatter plot of X (spy) and Y (ko) to examine the nature of their relationship visually

plt.figure(figsize = (10, 6))
plt.rcParams.update({'font.size': 14})
plt.xlabel("Sugars")
plt.ylabel("Ratings")
plt.title("Scatter plot of ratings ")
plt.scatter(df['SUGARS'], df['RATING'])
abline(-2.4008, 59.2844)
plt.show()
```

```python
#%% read from csv file
df = pd.read_csv('final_2022.csv')  
print(df) 
#%%
####################################################
## A scatter plot of X (spy) and Y (ko) to examine the nature of their relationship visually

plt.figure(figsize = (10, 6))
plt.rcParams.update({'font.size': 14})
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot of X-Y ")
plt.scatter(df['X'], df['Y'])
abline(1.85, 1.4)
plt.show()
```

```python
## 2a. Fit a simple linear regression model to the data using statsmodels 

### Create an instance of the class OLS
slr_sm_model = smf.ols('Y ~ X', data=df)

### Fit the model (statsmodels calculates beta_0 and beta_1 here)
slr_sm_model_rating = slr_sm_model.fit()

### Summarize the model

print(slr_sm_model_rating.summary()) 
fig = plt.figure(figsize=(8, 6))
param_slr = slr_sm_model_rating.params
sm.graphics.plot_regress_exog(slr_sm_model_rating, "X", fig=fig)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np    

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
#    axes = plt.gca()
#    x_vals = np.array(axes.get_xlim())
    x_vals = np.linspace(0,14, 100, endpoint=True)
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
```
