```python
BMI = float(input("請輸入身體質量指數(BMI值):"))
if BMI >= 35:
    print("重度肥胖!")
```

```python
temperature = float(input("請輸入身體溫度(攝氏°C):"))
if temperature >= 37.5:
    print("屬於發燒!!")
else:
    print("屬於正常體溫~")
```
